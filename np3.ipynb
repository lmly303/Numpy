```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Zomato sales analysis ===")
print("shape of the data : ", sales_data.shape)
print("first 3 restraunt data : ", sales_data[0:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))

# yearly total
yearly_total = np.sum(sales_data[:, 1:],axis=0)
print(yearly_total)

# min sales per restraunt
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# average sales per restraunt
avg_sales = np.mean(sales_data[:, 1:],axis=1)
print(avg_sales)

# max sales per restraunt
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)
```

```python
# cumulitive sum 1, 2, 3 = 1, 3, 6
cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cum_sales)

plt.figure(figsize=(10,6))
# figure(): This function creates a new figure (i.e., a blank canvas for plotting).
# figsize=(10,6): This sets the size of the figure in inches:
# 10 inches wide
# 6 inches tall
plt.plot(np.mean(cum_sales, axis=0))
plt.title("average cumulative sales across all restraunts")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
